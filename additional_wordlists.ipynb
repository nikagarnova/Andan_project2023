### Собираем список бесполезных слов

```python
df = pd.read_csv('/content/result.csv')

# Создаем список уникальных жанров
genres = set()
for genre_list in df['genre']:
    genres.update(genre_list.split(', '))

# Создаем новые столбцы для каждого жанра и заполняем их нулями
for genre in genres:
    df[genre] = 0

# Обновляем значения столбцов жанров
for index, row in df.iterrows():
    genre_list = row['genre'].split(', ')
    for genre in genre_list:
        df.at[index, genre] = 1

films = df['description']
descriptions = ' '.join(films)
tokens = word_tokenize(descriptions)
freq_dist = FreqDist(tokens)

# Получение 500 наиболее часто встречающихся слов и их количеств
top_500_words = freq_dist.most_common(500)

# Вывод результата
for word, count in top_500_words:
    print(word,count)
```
